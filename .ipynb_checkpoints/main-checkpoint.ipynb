```python
# Dependencies
import tweepy
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from config import (consumer_key,
                    consumer_secret,
                    access_token,
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

```python
news_users = ("@nytimes", "@cnn", "@BBCWorld", "@CBSNews", "@FoxNews")
```

```python
user_results = []

counter = 0
oldest_tweet = None

# Loop through each user
for user in news_users:
    counter = 0
    oldest_tweet = None

    # Loop through 5 pages of tweets (total 200 tweets)
    for x in range(1, 6):

        # Get all tweets from home feed
        public_tweets = api.user_timeline(user, page=x, max_id=oldest_tweet)

        # Loop through all tweets
        for tweet in public_tweets:

            # Run Vader Analysis on each tweet
            results = analyzer.polarity_scores(tweet["text"])
            compound = results["compound"]
            pos = results["pos"]
            neu = results["neu"]
            neg = results["neg"]
            tweet_text = tweet['text']
            
            # Add each value to the appropriate list
            date = tweet["created_at"]
            converted_time = datetime.strptime(date, "%a %b %d %H:%M:%S %z %Y")
        
        # Create a dictionaty of results
            user_results.append({
            "User": user,
            "Date": converted_time,
            "Tweet": tweet_text,
            "Compound Score": compound,
            "Positive Score": pos,
            "Neutral Score": neu,
            "Negative Score": neg,
            "Tweet Count": counter
                })
            counter += 1
```

```python
df = pd.DataFrame(user_results)
df = df[["User",
         "Date",
         "Tweet Count",
         "Tweet",
         "Compound Score", 
         "Positive Score", 
         "Neutral Score", 
         "Negative Score"]]
df = df.sort_values("Tweet Count", ascending=False)
df.to_csv("newssentiment.csv")
```

```python
df_nytimes = df[df["User"] == "@nytimes"]
df_cnn = df[df["User"] == "@cnn"]
df_bbcworld = df[df["User"] == "@BBCWorld"]
df_cbsnews = df[df["User"] == "@CBSNews"]
df_foxnews = df[df["User"] == "@FoxNews"]
```

```python
plt.scatter(df_nytimes["Tweet Count"], df_nytimes["Compound Score"], edgecolors="black")
plt.scatter(df_cbsnews["Tweet Count"], df_cbsnews["Compound Score"], edgecolors="black")
plt.scatter(df_cnn["Tweet Count"], df_cnn["Compound Score"], edgecolors="black")
plt.scatter(df_bbcworld["Tweet Count"], df_bbcworld["Compound Score"], edgecolors="black")
plt.scatter(df_foxnews["Tweet Count"], df_foxnews["Compound Score"], edgecolors="black")
plt.xlabel("Tweets Ago")
plt.ylabel("Tweet Polarity")
plt.title("Sentiment Analysis of Media Tweets (07/31/2018)")
plt.legend(news_users, bbox_to_anchor =(1,1), title="Media Sources")
sns.set()

plt.savefig("tweetpolarityscatter.png", dpi=100,  bbox_inches='tight')
plt.show()
```

```python
average_group = df.groupby("User").mean()["Compound Score"]
averages = []

for x in average_group:
    averages.append(x)
```

```python
colors = ["skyblue", "green", "red", "blue", "yellow"]
sns.set()
average_group.plot.bar(color=colors, align = "center", width=1, edgecolor = "black", )

plt.title("Overall Media Sentiment based on Twitter (07/31/2018)")
plt.xlabel("")
plt.xticks(np.arange(len(news_users)), ("BBC", "CBS", "CNN", "Fox", "NYT"), rotation=0)
plt.ylabel("Tweet Polarity")

count = 0
for x in averages:
    if x < 0:
        ybar = max(averages)*.10
    else:
        ybar = min(averages)*.10
    plt.text(count, ybar, round(x,2), ha="center")
    count += 1
    
plt.grid(None)
plt.savefig("tweetpolarity.png", dpi=100)

plt.show()
```
